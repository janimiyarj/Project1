```python
import pandas as pd
import sweetviz as sv
```

```python
df = pd.read_csv('/content/2017_Yellow_Taxi_Trip_Data.csv')
```

```python
df.head()
```

```python
my_report = sv.analyze([df, "Train"],target_feat='fare_amount')
```

```python
# show the report in a form of an HTML file
my_report.show_html('My_report.html')
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, train_size=0.75)
# compare the dataset
compare = sv.compare(source=train_df, compare=test_df, target_feat="fare_amount")

# Show the result
compare.show_html('Compare.html')
```
